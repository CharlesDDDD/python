```python
n1,n2,n3=432,232,336
p,q=0.5,0.5

for i in range(100):
    p_new=(((p*n1)/(p+q-p*q))+((p*n2)/(1-q+p*q)))/(n1+n2+n3)
    q_new=(n3+2*n1-(p*n1/(p+q-p*q)))/(2*(n1+n2+n3)-(p*n1/(p+q-p*q))-(p*n2/(1-q+p*q)))
    print('第'+str(i+1)+'次迭代之后p值为: ',p_new)
    print('第'+str(i+1)+'次迭代之后q值为: ',q_new)
    p=p_new
    q=q_new

print(p_new,q_new)
```

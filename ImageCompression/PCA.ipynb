```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def comp_2d(image_2d,mean_mat):
    #print(image_2d)
    cov_m = image_2d - mean_mat #np.mean(image_2d,axis = 0)
    #print(cov_m)
    eig_val,eig_vec = np.linalg.eig(np.cov(cov_m))
    
    idx = np.argsort(eig_val)
    idx = idx[::-1]
    eig_vec = eig_vec[:,idx]
    eig_val = eig_val[idx]
    
    #选取特征值个数
    eig_val_sum = np.sum(eig_val)
    r = np.linalg.matrix_rank(image_2d)
    p_eig = 0
    for i in range(r):
        p_eig = p_eig + eig_val[i]
        if p_eig > 0.99*eig_val_sum :
            break
    n_pick = i
    eig_vec =eig_vec[:,range(n_pick)]
    
    #计算压缩率
    score = np.dot(eig_vec.T,cov_m)
    #计算重构误差
    recon = np.dot(eig_vec,score) + mean_mat#+ np.mean(image_2d,axis = 0).T
    recon_img_mat = np.uint8(np.absolute(recon))
    
    compress = np.size(eig_vec)+ np.size(score)#+np.size(np.mean())
    origin_img = np.size(image_2d) 
    error = np.linalg.norm(recon_img_mat-image_2d)
    return recon_img_mat,compress,origin_img,error

    
    
    
    
```
